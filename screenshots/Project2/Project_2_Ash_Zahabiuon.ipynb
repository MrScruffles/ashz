**Loading the Dataset**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Read the dataset
df = pd.read_csv('/kaggle/input/apple-stock-data-updated-till-22jun2021/AAPL.csv')

# Convert 'Date' to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Proceed with data exploration and analysis
# Create new columns for 'Year', 'Month', 'Day', 'DayOfWeek'
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['DayOfWeek'] = df['Date'].dt.dayofweek

# Calculate daily returns
df['Daily Return'] = df['Adj Close'].pct_change()

# Proceed with your transformations
df['Price Change'] = df['Adj Close'].diff()
df['Price Movement'] = df['Price Change'].apply(lambda x: 1 if x > 0 else (-1 if x < 0 else 0))

# Display the updated DataFrame
df.head()
```

**Model / Training**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

# Read the dataset
# Create a new column to categorize the price change between consecutive days
df['Price Change'] = df['Adj Close'].diff()

# Create a classification label based on the price change:
# - Increase (1): Price increased from the previous day
# - Decrease (-1): Price decreased from the previous day
# - No Change (0): Price stayed the same
df['Price Movement'] = df['Price Change'].apply(lambda x: 1 if x > 0 else (-1 if x < 0 else 0))

# Calculate some features: 7-day and 30-day moving averages for the adjusted close price
df['7_day_MA'] = df['Adj Close'].rolling(window=7).mean()
df['30_day_MA'] = df['Adj Close'].rolling(window=30).mean()

# Let's also include the volume as a feature for the model
df = df[['Adj Close', 'Volume', '7_day_MA', '30_day_MA', 'Price Movement']].dropna()

# Define features (X) and labels (y)
X = df[['Adj Close', 'Volume', '7_day_MA', '30_day_MA']]
y = df['Price Movement']

# Split the dataset into training and testing sets (75% training, 25% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Initialize the Multinomial Naive Bayes model
clf = MultinomialNB()

# Train the model
clf.fit(X_train, y_train)

# Predict on the test set
y_pred = clf.predict(X_test)

# Display the classification report to evaluate the model
print(classification_report(y_test, y_pred, target_names=['Decrease', 'No Change', 'Increase']))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Generate confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
plt.figure(figsize=(8,6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=['Decrease', 'No Change', 'Increase'], 
            yticklabels=['Decrease', 'No Change', 'Increase'])
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Plot the distribution of actual vs predicted
plt.figure(figsize=(10,6))
plt.hist([y_test, y_pred], label=['Actual', 'Predicted'], color=['blue', 'orange'], alpha=0.7, bins=3)
plt.xticks(ticks=[-1, 0, 1], labels=['Decrease', 'No Change', 'Increase'])
plt.title('Actual vs Predicted Class Distribution')
plt.xlabel('Price Movement')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Initialize the K-Nearest Neighbors model
knn_model = KNeighborsClassifier(n_neighbors=5)

# Train the model
knn_model.fit(X_train, y_train)

# Predict on the test set
y_pred_knn = knn_model.predict(X_test)

# Display the classification report
print("Classification Report for K-Nearest Neighbors (KNN):\n")
print(classification_report(y_test, y_pred_knn, target_names=['Decrease', 'No Change', 'Increase']))
```

```python
# Preprocess the dataset by adding necessary features for classification
df['Price Change'] = df['Adj Close'].diff()

# Create a classification label based on the price change
df['Price Movement'] = df['Price Change'].apply(lambda x: 1 if x > 0 else (-1 if x < 0 else 0))

# Calculate moving averages
df['7_day_MA'] = df['Adj Close'].rolling(window=7).mean()
df['30_day_MA'] = df['Adj Close'].rolling(window=30).mean()

# Clean up the dataset, dropping rows with NaN values from rolling averages and price changes
df = df[['Adj Close', 'Volume', '7_day_MA', '30_day_MA', 'Price Movement']].dropna()

# Define features (X) and labels (y)
X = df[['Adj Close', 'Volume', '7_day_MA', '30_day_MA']]
y = df['Price Movement']

# Split the dataset for training and testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# Now let's generate the learning curve plot for Random Forest
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import learning_curve
from sklearn.model_selection import StratifiedKFold
import matplotlib.pyplot as plt

# Define the Random Forest classifier
model = RandomForestClassifier(random_state=42)

# Generate the learning curve data
train_sizes, train_scores, test_scores = learning_curve(
    model, X_train, y_train, cv=StratifiedKFold(n_splits=5), n_jobs=-1,
    train_sizes=[0.1, 0.2, 0.5, 0.7, 1.0], scoring='accuracy'
)

# Calculate mean and standard deviation for plotting
train_mean = train_scores.mean(axis=1)
test_mean = test_scores.mean(axis=1)
train_std = train_scores.std(axis=1)
test_std = test_scores.std(axis=1)

# Plot the learning curve
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, 'o-', color="r", label="Training score")
plt.plot(train_sizes, test_mean, 'o-', color="g", label="Cross-validation score")

# Plot the std deviation as a shaded region
plt.fill_between(train_sizes, train_mean - train_std, train_mean + train_std, color="r", alpha=0.2)
plt.fill_between(train_sizes, test_mean - test_std, test_mean + test_std, color="g", alpha=0.2)

# Add labels and title
plt.title("Learning Curve (Random Forest)")
plt.xlabel("Training Set Size")
plt.ylabel("Accuracy Score")
plt.legend(loc="best")

# Show the plot
plt.grid()
plt.show()
```
